```python
#create train/val/test splits at 80%/10%/10% respectively
#test set should contain hands that are never seen in training!
#one time run. Creates train/validation/test folders in root folder for our implementation in google drive.
import glob
import random
import os
import shutil
random.seed(100) #seed so that rerunning will not change data in the folders
root_path = "/content/drive/MyDrive/Lab3_Gestures_W21/"
#create list of students
list_of_students = list(set([i.lstrip(root_path + "A/").split("_")[0] for i in glob.glob(root_path + "A/*.jpg")]))
total_students = len(list_of_students)
#choose students for training, testing, and validation sets
train_students = []
val_students = []
test_students = []
for i in range(int(total_students*0.8)):
  train_choice = random.choice(list_of_students)
  train_students.append(train_choice)
  list_of_students.remove(train_choice)
for i in range(int(total_students*0.1)):
  val_choice = random.choice(list_of_students)
  val_students.append(val_choice)
  list_of_students.remove(val_choice)
test_students = list_of_students #since only 10% test data is left
#create the folders
classes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
for letter in classes:
  os.makedirs(os.path.join(root_path,"train",letter),exist_ok=True)
  os.makedirs(os.path.join(root_path,"val",letter),exist_ok=True)
  os.makedirs(os.path.join(root_path,"test",letter),exist_ok=True)
  for train_student in train_students:
    for train_img in glob.glob(root_path + letter + "/" + train_student + "*"):
      shutil.copy(train_img, os.path.join(root_path, "train", letter))
  for val_student in val_students:
    for val_img in glob.glob(root_path + letter + "/" + val_student + "*"):
      shutil.copy(val_img, os.path.join(root_path, "val",letter))  
  for test_student in test_students:
    for test_img in glob.glob(root_path + letter + "/" + test_student + "*"):
      shutil.copy(test_img, os.path.join(root_path, "test",letter))  
```

```python
def data_loader(batch_size = 32, mode = "train", root_path = "/content/"):
  transform = transforms.Compose([transforms.Resize((224,224)), transforms.ToTensor(), transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
  if mode == 'train':
    dataset = torchvision.datasets.ImageFolder(root_path + "train/", transform=transform)
    dataloader = DataLoader(dataset, batch_size=batch_size, num_workers = 2, shuffle=True)
  if mode == 'val':
    dataset = torchvision.datasets.ImageFolder(root_path + "val/", transform=transform)
    dataloader = DataLoader(dataset, batch_size=batch_size, num_workers = 2, shuffle=True)
  if mode == 'test':
    dataset = torchvision.datasets.ImageFolder(root_path + "test/", transform=transform)     
    dataloader = DataLoader(dataset, batch_size=batch_size, num_workers = 2, shuffle=True)
  return dataloader
```

### Part (a) Convolutional Network - 5 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
class CNN(nn.Module):
  def __init__(self):
    self.name = "CNN"
    super(CNN, self).__init__()
    #here i create the 5 conv modules
    #using sequential allows me to better visaulise the model instead of using these layers in the forward function.
    self.Conv1 = nn.Sequential(
    nn.Conv2d(3, 32, 5), # 220, 220
    nn.MaxPool2d(2), # 110, 110
    nn.ReLU(),
    nn.BatchNorm2d(32)
    )
    self.Conv2 = nn.Sequential(
    nn.Conv2d(32, 64, 5), # 106, 106
    nn.MaxPool2d(2),  # 53,53
    nn.ReLU(),
    nn.BatchNorm2d(64)
    )
    self.Conv3 = nn.Sequential(
    nn.Conv2d(64, 128, 3), # 51, 51
    nn.MaxPool2d(2), # 25, 25
    nn.ReLU(),
    nn.BatchNorm2d(128)
    )
    self.Conv4 = nn.Sequential(
    nn.Conv2d(128, 256, 3), # 23, 23
    nn.MaxPool2d(2), # 11, 11
    nn.ReLU(),            
    nn.BatchNorm2d(256)
    )
    self.Conv5 = nn.Sequential(
    nn.Conv2d(256, 512, 3), # 9, 9
    nn.MaxPool2d(2), # 4, 4
    nn.ReLU(),
    nn.BatchNorm2d(512)
    )
    
    self.Linear1 = nn.Linear(512 * 4 * 4, 256)
    self.dropout = nn.Dropout(0.1)
    self.Linear3 = nn.Linear(256, 9)

  def forward(self, x):
    x = self.Conv1(x)
    x = self.Conv2(x)
    x = self.Conv3(x)
    x = self.Conv4(x)
    x = self.dropout(x)
    x = self.Conv5(x)
    x = x.view(x.size(0), -1)
    x = self.Linear1(x)
    x = self.dropout(x)
    x = self.Linear3(x)
    return x
```

The model consists of 5 conv modules, each containing:

1.   2D conv layer which uses a 32 channeled kernel of size (5,5,32) to convert the image from (224,224,3) --> (220, 220,32). 
2.   MaxPool2d converts the image to (110, 110, 32).
3. ReLU is used for activation. Size remains the same. 
4. Batch Normalization is used for nomralisation, size remains the same. 

This process is done 5 times with reduced kernel size of (3,3) at the third conv module, such that the image goes from (224, 224, 32) --> (110, 110, 32) --> (106, 106, 64) --> (53, 53, 64) --> (51, 51, 128) --> (25, 25, 128) --> (23, 23, 256) --> (11, 11, 256) --> (9, 9, 512) --> (4, 4, 512).

After this, the output is flattened from (4, 4, 512) to 256, after which a dropout layer with 10% probability is added. Finally, a second linear layer converts the result to 9 output, which represents the number of classes.


The architecture uses 5 layers as it is deep enough for accurate image classification but still parameter efficient and not as big as models such as ResNet-50 and others. CNN layers are used, as they are locally connected layers which have local correlations and learn to detect features.

Fully connected layers are used at the end to use the detected features from the CNNs and convert them into a class probability. Pooling layers are used to reduce the size and keep important information while removing the unnecessary information. MaxPool is used for this purpose. Lastly, ReLU is used for activation as lab 3a has shown it works as a great activation layer for image classification tasks. 

### Part (b) Training Code - 5 pt

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

```python
# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
def get_accuracy(model, loader):   
  model.eval()
  correct_pred, num_examples = 0, 0
  with torch.no_grad():
    for imgs, labels in loader:
      imgs = imgs.to(device)
      labels = labels.to(device)
      outputs = model(imgs)
      predicted = torch.softmax(outputs,dim=1)
      _, predicted_labels = torch.max(predicted, 1)
      num_examples += labels.size(0)
      correct_pred += (predicted_labels == labels).sum()
  return correct_pred.float()/num_examples * 100
```

```python
# Trainer Helper - for saving data
def get_model_name(batch_size, learning_rate, epoch):
    """ Generate a name for the model consisting of all the hyperparameter values

    Args:
        config: Configuration object containing the hyperparameters
    Returns:
        path: A string with the hyperparameter name and value concatenated
    """
    path = "bs{}_lr{}_epoch{}".format(batch_size, learning_rate, epoch)
    return path
```

```python
# Training Model - For CNN Training

def train(model, batch_size, learning_rate, num_epochs=1, root_path = "/content/"):
  model = model.to(device)
  train_loader = data_loader(batch_size = batch_size, root_path = root_path)
  val_loader = data_loader(batch_size = batch_size, mode = "val", root_path = root_path)
  
  torch.manual_seed(100) # set the random seed
  
  # Define the Loss function and optimizer
  criterion = nn.CrossEntropyLoss().to(device) #ideal loss function for multi-variate classification
  optimizer = optim.Adam(model.parameters(), lr=learning_rate)
  scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=3, factor=0.7, verbose= True, min_lr=1e-6)
  iters, losses, train_acc, val_acc = [], [], [], []
  
  #create backup folder
  model_name = get_model_name(batch_size, learning_rate, num_epochs)
  model_path = os.path.join("/content/drive/MyDrive/Backup", model_name)
  os.makedirs(model_path, exist_ok=True)

  # training
  start_time = time.time() 
  for epoch in range(num_epochs):
    model.train()
    for i, (images,labels) in enumerate(train_loader):
      images = images.to(device)
      labels = labels.to(device)
            
      logits = model(images)
      loss=criterion(logits, labels)
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

    if epoch % 5 == 0: # accuracy checked every 5 laps to improve training performance
      iters.append(epoch)
      losses.append(float(loss))  
      train_acc.append(float(get_accuracy(model, train_loader))) # compute training accuracy 
      val_acc.append(float(get_accuracy(model, val_loader)))  # compute validation accuracy

      print("Epoch [{}/{}], Loss: {:.4f}, Training accuracy: {:.4f}, Validation Accuracy: {:.4f}".format(epoch+1,num_epochs,loss,train_acc[-1],val_acc[-1]))
      scheduler.step(val_acc[-1])
      try: 
        if val_acc[-1]==max(val_acc): #model saved if performance is better 
          torch.save(model.state_dict(), model_path + "/" + "best")
          print("best model saved at epoch {}".format(epoch+1))
      except:
        continue
    if epoch%20 == 0: #model checkpoint every 20 epochs
      torch.save(model.state_dict(), model_path + "/" + get_model_name(batch_size, learning_rate, epoch))
      print ("checkpoint saved at epoch {}".format(epoch+1))      

      

  print("Total time taken: {}".format(time.time() - start_time))   

  # plotting
  plt.title("Training Curve")
  plt.plot(iters, losses, label="Train")
  plt.xlabel("Iterations")
  plt.ylabel("Loss")
  plt.show()
  plt.savefig(str(model_path) + '/training curve.png')

  plt.title("Validation Curve")
  plt.plot(iters, train_acc, label="Train")
  plt.plot(iters, val_acc, label="Validation")
  plt.xlabel("Iterations")
  plt.ylabel("Training Accuracy")
  plt.legend(loc='best')
  plt.show()
  plt.savefig(str(model_path) + '/validation curve.png')
  
  print("Final Training Accuracy: {}".format(train_acc[-1]))
  print("Final Validation Accuracy: {}".format(val_acc[-1]))
```

The loss function used is cross entropy loss, as we are trying to classify images into multiple classes.

Adam Optimizer is used as it incorporates momentum an d adaptive learning rate while requiring minimal tuning, leading to a rapid convergence.

```python
import torchvision.models
alexnet = torchvision.models.alexnet(pretrained=True)
```

The alexnet model is split up into two components: *alexnet.features* and *alexnet.classifier*. The
first neural network component, *alexnet.features*, is used to compute convolutional features, which are
taken as input in *alexnet.classifier*.

The neural network alexnet.features expects an image tensor of shape Nx3x224x224 as input and it will
output a tensor of shape Nx256x6x6 . (N = batch size).

Compute the AlexNet features for each of your training, validation, and test data. Here is an example code
snippet showing how you can compute the AlexNet features for some images (your actual code might be
different):

```python
# img = ... a PyTorch tensor with shape [N,3,224,224] containing hand images ...
def alexnet_features(batch_size, root_path, save_path):
  train_loader = data_loader(batch_size = batch_size, root_path = root_path)
  val_loader = data_loader(batch_size = batch_size, mode = "val", root_path = root_path)
  test_loader = data_loader(batch_size = batch_size, mode = "test", root_path = root_path)
  classes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
  for i,(img,label) in enumerate(train_loader):
    features = alexnet.features(img)
    feature_tensor = torch.from_numpy(features.detach().numpy())
    folder_name = os.path.join(save_path, "train", str(classes[label]))
    os.makedirs(folder_name, exist_ok=True)
    #print(str(classes[label]))
    #print(folder_name)
    torch.save(feature_tensor.squeeze(0), str(folder_name) +"/"+ str(i) + '.tensor')
  for i,(img,label) in enumerate(val_loader):
    features = alexnet.features(img)
    feature_tensor = torch.from_numpy(features.detach().numpy())
    folder_name = os.path.join(save_path, "val", str(classes[label]))
    os.makedirs(folder_name, exist_ok=True)
    #print(str(val_loader.dataset.samples[i][0]).split("/")[-1].rstrip(".jpg"))
    torch.save(feature_tensor.squeeze(0), str(folder_name) +"/"+ str(i) + '.tensor')
  for i,(img,label) in enumerate(test_loader):
    features = alexnet.features(img)
    feature_tensor = torch.from_numpy(features.detach().numpy())
    folder_name = os.path.join(save_path, "test", str(classes[label]))
    os.makedirs(folder_name, exist_ok=True)
    #print(str(val_loader.dataset.samples[i][0]).split("/")[-1].rstrip(".jpg"))
    torch.save(feature_tensor.squeeze(0), str(folder_name) +"/"+ str(i) + '.tensor')

alexnet_features(1, '/content/', '/content/alexnet_features/')
```

### Part (b) - 3 pt
Build a convolutional neural network model that takes as input these AlexNet features, and makes a
prediction. Your model should be a subclass of nn.Module.

Explain your choice of neural network architecture: how many layers did you choose? What types of layers
did you use: fully-connected or convolutional? What about other decisions like pooling layers, activation
functions, number of channels / hidden units in each layer?

Here is an example of how your model may be called:

```python
class alex_pred(nn.Module):
  #as this is the output from alexnet_features resembles the output from my 4th conv module, I use the 5th conv module without the pooling and the fully connected layers here.

  def __init__(self):
    self.name = "alex_pred"
    super(alex_pred, self).__init__()
    self.Conv5 = nn.Sequential(
    nn.Conv2d(256, 512, 3),
    nn.ReLU(),
    nn.BatchNorm2d(512)
    )
    
    self.Linear1 = nn.Linear(512 * 4 * 4, 256)
    self.dropout = nn.Dropout(0.5)
    self.Linear3 = nn.Linear(256, 9)

  def forward(self, x):
    x = self.Conv5(x)
    x = x.view(x.size(0), -1)
    x = self.Linear1(x)
    x = self.dropout(x)
    x = self.Linear3(x)
    return x
```

```python
# Load data 
alex_train_folder = torchvision.datasets.DatasetFolder("/content/alexnet_features/train/", loader=torch.load, extensions=('.tensor'))
alex_val_folder = torchvision.datasets.DatasetFolder("/content/alexnet_features/val/", loader=torch.load, extensions=('.tensor'))
alex_test_folder = torchvision.datasets.DatasetFolder("/content/alexnet_features/test/", loader=torch.load, extensions=('.tensor'))
```

```python
def train(model, batch_size, learning_rate, num_epochs=1):
  model = model.to(device)
  train_loader = torch.utils.data.DataLoader(alex_train_folder, batch_size=batch_size, shuffle=True, drop_last = True)
  val_loader = torch.utils.data.DataLoader(alex_val_folder, batch_size=batch_size, shuffle=True, drop_last = True)
  torch.manual_seed(100) # set the random seed
  
  # Define the Loss function and optimizer
  criterion = nn.CrossEntropyLoss().to(device) #ideal loss function for multi-variate classification
  optimizer = optim.Adam(model.parameters(), lr=learning_rate)
  scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5, verbose= True, min_lr=1e-6)
  iters, losses, train_acc, val_acc = [], [], [], []
  
  #create backup folder
  model_name = get_model_name(batch_size, learning_rate, num_epochs) 
  model_path = os.path.join("/content/drive/MyDrive/Backup/AlexNet/", model_name)
  os.makedirs(model_path, exist_ok=True)

  # training
  start_time = time.time() 
  for epoch in range(num_epochs):
    model.train()
    for i, (images,labels) in enumerate(train_loader):
      images = images.to(device)
      labels = labels.to(device)
            
      logits = model(images)
      loss=criterion(logits, labels)
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

  
    if epoch % 2 == 0:
      iters.append(epoch)
      losses.append(float(loss))  
      train_acc.append(float(get_accuracy(model, train_loader))) # compute training accuracy 
      val_acc.append(float(get_accuracy(model, val_loader)))  # compute validation accuracy

      print("Epoch [{}/{}], Loss: {:.4f}, Training accuracy: {:.4f}, Validation Accuracy: {:.4f}".format(epoch+1,num_epochs,loss,train_acc[-1],val_acc[-1]))
      scheduler.step(val_acc[-1])
      try: 
        if val_acc[-1]==max(val_acc): #model saved if performance is better 
          torch.save(model.state_dict(), model_path + "/" + "best")
          print("best model saved at epoch {}".format(epoch+1))
      except:
        continue
    if epoch%10 == 0:
      torch.save(model.state_dict(), model_path + "/" + get_model_name(batch_size, learning_rate, epoch))
      print ("checkpoint saved at epoch {}".format(epoch+1))      

  print("Total time taken: {}".format(time.time() - start_time))   

  # plotting
  plt.title("Training Curve")
  plt.plot(iters, losses, label="Train")
  plt.xlabel("Iterations")
  plt.ylabel("Loss")
  plt.show()
  plt.savefig(str(model_path) + '/training curve.png')

  plt.title("Validation Curve")
  plt.plot(iters, train_acc, label="Train")
  plt.plot(iters, val_acc, label="Validation")
  plt.xlabel("Iterations")
  plt.ylabel("Training Accuracy")
  plt.legend(loc='best')
  plt.show()
  plt.savefig(str(model_path) + '/validation curve.png')
  
  print("Final Training Accuracy: {}".format(train_acc[-1]))
  print("Final Validation Accuracy: {}".format(val_acc[-1]))
```
